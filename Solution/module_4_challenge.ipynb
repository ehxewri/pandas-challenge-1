## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
client_data_df['subtotal'] = client_data_df.apply((lambda x: x['unit_price']*x['qty']),axis=1)
print (client_data_df[['unit_price','qty','subtotal']].head(2).to_string(index=False),'\n')
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
client_data_df['shipping_weight'] = client_data_df.apply(lambda row: row['unit_weight'] * row['qty'],axis=1)
client_data_df['shipping_cost']   = client_data_df.apply(lambda row: 7.00 * row['shipping_weight'] 
                                    if row['unit_weight'] * row['qty'] > 50
                                    else 10.00 * row['unit_weight'] * row['qty'],axis=1)
print(client_data_df[['unit_price','unit_weight', 'qty', 'shipping_weight','shipping_cost']].head(3).to_string(index=False))
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

sales_tax = 1.0925
client_data_df['line_total'] = round(client_data_df.apply(lambda row: ((row['subtotal']) + row['shipping_cost'])*sales_tax,axis=1),2)
print(client_data_df[['qty', 'subtotal','shipping_cost','line_total']].head(3))
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

client_data_df['line_cost'] = client_data_df.apply(lambda row: row['unit_cost'] * row['qty'] + row['shipping_cost'], axis=1)
client_data_df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price
client_data_df['line_profit'] = client_data_df.apply(lambda row: row['line_total'] -row['line_cost'] , axis=1)

print(pd.concat([client_data_df['order_id'], client_data_df[['line_cost', 'line_total', 'line_profit']]
                 .map(format_currency)], axis=1).head().to_string(index=False))
```

```python
top_client_ids = client_data_df.groupby(['client_id']).size().nlargest(5).index.to_list()
for client_id in top_client_ids:
    # Calculate the sum
    customer_order_total = client_data_df.loc[client_data_df['client_id'] == client_id]['line_total'].sum()
    # format and print results
    print (f'Client {client_id}\tTotal\t${customer_order_total:.2f}')
```

Create a summary DataFrame showing the totals for the
for the top 5 clients with the following information:
total units purchased, total shipping price,
total revenue, and total profit. Sort by total profit.

```python
# Create an empty summary DataFrame with the specified columns and data types
summary_df = pd.DataFrame(columns=['client_id', 'total_qty', 'total_shipping', 'total_revenue', 'total_cost', 'total_profit'])

# Iterate over the top 5 client IDs
for client_id in top_client_ids:
    # Calculate totals for the current client
    total_qty      = client_data_df.loc[client_data_df['client_id'] == client_id]['qty'].sum()
    total_shipping = client_data_df.loc[client_data_df['client_id'] == client_id]['shipping_cost'].sum()
    total_revenue  = client_data_df.loc[client_data_df['client_id'] == client_id]['line_total'].sum()
    total_cost     = client_data_df.loc[client_data_df['client_id'] == client_id]['line_cost'].sum()
    total_profit   = client_data_df.loc[client_data_df['client_id'] == client_id]['line_profit'].sum()
    
    # Create a new DataFrame for the current client
    new_row_df = pd.DataFrame({'client_id'     : [client_id],
                               'total_qty'     : [total_qty],
                               'total_shipping': [total_shipping],
                               'total_revenue' : [total_revenue],
                               'total_cost'    : [total_cost],
                               'total_profit'  : [total_profit]})
    
    # Concatenate the new DataFrame with the summary DataFrame
    summary_df = pd.concat([summary_df, new_row_df], ignore_index=True, axis=0)

# Sort the summary DataFrame by total_profit in descending order
summary_df = summary_df #.sort_values(by='total_profit', ascending=False)

# Print the sorted summary DataFrame
summary_df.sort_values(by="total_profit",ascending=False)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df=summary_df.rename(columns=
    {'client_id':'Client ID', 
     'total_qty':'Units', 
     'total_shipping':'Shipping', 
     'total_revenue':'Total Revenue',
     'total_cost':'Total Cost', 
     'total_profit':'Total Profit'
})
# summary_df[['Client ID','Units','Shipping','Total Revenue','Total Cost','Total Profit']]
display_df = summary_df.assign(**{
    'Shipping'     : lambda x: ('$' + (x['Shipping']      / 1_000_000).round(2).astype(str) + 'M'),
    'Total Revenue': lambda x: ('$' + (x['Total Revenue'] / 1_000_000).round(2).astype(str) + 'M'),
    'Total Cost'   : lambda x: ('$' + (x['Total Cost']    / 1_000_000).round(2).astype(str) + 'M'),
    'Total Profit' : lambda x: ('$' + (x['Total Profit']  / 1_000_000).round(2).astype(str) + 'M')
})
display_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
display_df.sort_values(by="Total Profit",ascending=False)
```
